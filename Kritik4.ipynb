```python
import matplotlib.pyplot as plt
import numpy as np

# Function to perform gradient descent
def gradient_descent(f, learning_rate, initial_point):
    # Function to estimate the derivative of f at a given point using symmetric approximation
    def deriv(f, base_point):
        return (f(base_point + 10**(-10)) - f(base_point - 10**(-10))) / (2 * 10**(-10))

    # Initialize lists to store the sequence of x and f(x) values
    x_coords = [initial_point]  # Stores x_n values
    y_coords = [f(initial_point)]  # Stores f(x_n) values

    # Perform gradient descent for a fixed number of iterations
    for i in range(100):
        # Update x using the gradient descent formula
        x_coords.append(x_coords[-1] - learning_rate * deriv(f, x_coords[-1]))
        y_coords.append(f(x_coords[-1]))

    # Plotting the function and the sequence of points
    plot_range = np.linspace(min(x_coords) - 0.5, max(x_coords) + 0.5, 10000)  # Define a suitable range for plotting
    function_range = [f(i) for i in plot_range]  # Evaluate f(x) over the plot range
    plt.plot(plot_range, function_range, label="f(x)")  # Plot the function f(x)
    plt.plot(x_coords, y_coords, marker='o', label="Descent Path")  # Plot the gradient descent path
    plt.legend()
    plt.show()

    # Return the last x_n and f(x_n) values, rounded to 3 decimals
    return round(x_coords[-1], 3), round(y_coords[-1], 3)

# Define example functions
def f(x):
    return x**2

# Run gradient descent on f(x) with learning rate 0.8 and initial point 1
gradient_descent(f, 0.8, 1)

def g(x):
    return x**4 - 2 * x**2

# Run gradient descent on g(x) with learning rate 0.1 and initial point -0.5
gradient_descent(g, 0.1, -0.5)

def funny_function(x):
    # Define a piecewise function for the "funny function"
    if x > 0:
        return x**x
    elif x == 0:
        return 1
    else:
        return abs(x)**abs(x)

# Run gradient descent on the "funny function" with learning rate 0.2 and initial point -1
gradient_descent(funny_function, 0.2, -1)
```
