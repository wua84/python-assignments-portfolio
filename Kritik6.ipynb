```python
import numpy as np
import matplotlib.pyplot as plt

def normal_density(mean, variance, x):
    return (1 / (variance * np.sqrt(2 * np.pi))) * np.exp(-((x - mean) ** 2) / (2 * variance ** 2))

def integrate(a, b, f):
    n = 1000
    step = (b - a) / n
    area = 0
    for i in range(n):
        area += step * f(a + i * step)
    return area

def probability_between(mean, variance, a, b):
    def f(x):
        return normal_density(mean, variance, x)
    return integrate(a, b, f)

#values
mean = 171
variance = 7.1
a = 162
b = 190

#probability
probability = probability_between(mean, variance, a, b)
print("Probability:", probability)

x = np.linspace(mean - 4 * variance, mean + 4 * variance, 1000)
y = [normal_density(mean, variance, xi) for xi in x]

plt.plot(x, y, color="blue")

plt.xlabel("Height (cm)")
plt.ylabel("Density")
plt.title("Normal Distribution of Male Heights")
plt.show()

```

```python
import numpy as np

#density function
def normal_density(mean, variance, x):
    std_dev = np.sqrt(variance)
    return (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-((x - mean) ** 2) / (2 * variance))

#integration
def integrate(a, b, f):
    n = 1000
    step = (b - a) / n
    total = 0
    for i in range(n):
        total += step * f(a + i * step)
    return total

#expected value E[X] for a normal distribution
def expected_value_normal(mean, variance, a, b):
    def f(x):
        return x * normal_density(mean, variance, x)
    return integrate(a, b, f)

#values
mean_normal = 0
variance_normal = 1
E_X = expected_value_normal(mean_normal, variance_normal, -10, 10)
print("Expected value (Normal Distribution):", E_X)

def exponential_pdf(rate, x):
    if x >= 0:
        return rate * np.exp(-rate * x)
    else:
        return 0

def expected_value_exponential(rate, a, b):
    def f(x):
        return x * exponential_pdf(rate, x)
    return integrate(a, b, f)

lambda_exp = 1 / 50  # Given rate
E_X_exp = expected_value_exponential(lambda_exp, 0, 500)
print("Expected value (Exponential Distribution):", E_X_exp)

def expected_x_squared(mean, variance, a, b):
    def f(x):
        return x**2 * normal_density(mean, variance, x)
    return integrate(a, b, f)

mean_height = 171
variance_height = 7.1
E_X2 = expected_x_squared(mean_height, variance_height, 162, 190)

drug_dosage = 2.38 * E_X2
print("Estimated drug dosage:", drug_dosage)
```
