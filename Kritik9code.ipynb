```python
#(a)
import sympy as sp

# Define symbols
x, y = sp.symbols('x y')

# Define the function
f = sp.exp(x) * sp.sin(y) + y**3

# Compute partial derivatives
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print("∂f/∂x =", df_dx)
print("∂f/∂y =", df_dy)
```

```python
#(b)
# Define the function
g = x**2 * y + x * y**2

# Compute gradient (vector of partial derivatives)
grad_g = [sp.diff(g, var) for var in (x, y)]

# Evaluate gradient at (1, -1)
grad_at_point = [deriv.subs({x: 1, y: -1}) for deriv in grad_g]

# Compute magnitude of gradient
magnitude = sp.sqrt(sum(comp**2 for comp in grad_at_point))

print("Gradient vector:", grad_g)
print("Gradient at (1, -1):", grad_at_point)
print("Magnitude of gradient at (1, -1):", magnitude)
```

```python
# (c)
h = sp.log(x**2 + y**2)

# Compute second partial derivatives
d2h_dx2 = sp.diff(h, x, x)
d2h_dy2 = sp.diff(h, y, y)
d2h_dxdy = sp.diff(h, x, y)

print("∂²h/∂x² =", d2h_dx2)
print("∂²h/∂y² =", d2h_dy2)
print("∂²h/∂x∂y =", d2h_dxdy)

# Check if mixed partials are equal
print("Are mixed partials equal?", sp.simplify(d2h_dxdy - sp.diff(h, y, x)) == 0)
```

```python
# (d)create a contour plot of the function j(x,y) = x3−3xy+y3
import numpy as np
import matplotlib.pyplot as plt

# Define the function
x, y = sp.symbols('x y')
j = x**3 - 3*x*y + y**3

# Convert to numerical function
j_func = sp.lambdify((x, y), j, 'numpy')

# Create grid and evaluate
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

# Create plot
plt.figure(figsize=(10, 8))
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
#(e) Create another contour plot of a function of your choice.
import numpy as np
import matplotlib.pyplot as plt

# Define symbols
x, y = sp.symbols('x y')

k = x*sp.sin(y) + y*sp.cos(x) + 0.3*sp.sin(x + y)

# Convert to numerical function
k_func = sp.lambdify((x, y), k, 'numpy')

# Create grid
x_vals = np.linspace(-2.5, 2.5, 800)
y_vals = np.linspace(-2.5, 2.5, 800)
X, Y = np.meshgrid(x_vals, y_vals)

# Evaluate function
Z = k_func(X, Y)

# Create plot
plt.figure(figsize=(12, 10))
contour = plt.contourf(X, Y, Z, levels=100, cmap='seismic', vmin=-1.5, vmax=1.5)

# Add contour lines
plt.contour(X, Y, Z, levels=15, colors='black', linewidths=0.4, alpha=0.3)

# Formatting
plt.title(r'$k(x,y) = \sin(x^3 - y^2)\cos(x + y^3) + 0.2\sin(5x)\cos(5y)$', 
          pad=20, fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.colorbar(contour, label='Function Value')

# Show plot
plt.tight_layout()
plt.show()
```
